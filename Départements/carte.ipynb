```python
import json

with open('contour-des-departements.geojson') as mon_fichier:
    data = json.load(mon_fichier)
#data
```

```python
import matplotlib.pyplot as plt
import pandas as pd

fig, ax = plt.subplots()

df = pd.DataFrame()
n = len(data['features'])
for i in range(n):
    df = pd.DataFrame(data['features'][i]['geometry']['coordinates'][0])
    if df.shape[1]!=2:
        df = pd.DataFrame(data['features'][i]['geometry']['coordinates'][0][0])
    if df.shape[0]==0:
        print('ERROR')
    x = df.iloc[:,0]
    y = df.iloc[:,1]
    ax.plot(x,y, color = 'black')
```
